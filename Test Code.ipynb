## Using matplotlib stacked bar plot

```python
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
labels = ['G1','G2','G3','G4','G5']
x = np.arange(len(labels))
width = 0.4

fig,ax = plt.subplots()

reacts1 = ax.bar(x-width/2, men, width, label = 'men',linewidth = 1, color = 'orange', edgecolor = 'black')
reacts2 = ax.bar(x+width/2, women, width, label = 'women', linewidth = 1, edgecolor = 'black')

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

plt.tight_layout()
plt.legend(loc = (1.05,0.8))



#Auto Label on top of the bar plot
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(reacts1)
autolabel(reacts2)
```

```python
fig,ax = plt.subplots()

reacts1 = ax.bar(x, men, width, label = 'men',linewidth = 1, color = 'orange', edgecolor = 'black')
reacts2 = ax.bar(x, women, width, label = 'women', linewidth = 1, edgecolor = 'black', bottom=men)

ax.set_xticks(x)
ax.set_yticks(np.arange(0, 81, 10))
ax.set_xticklabels(labels)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

plt.legend(loc = (1.05,0.8))
```
