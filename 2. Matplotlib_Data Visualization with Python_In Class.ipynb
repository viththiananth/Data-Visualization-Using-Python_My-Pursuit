## create a bar plot of the following scores by group and gender. Use multiple X values on the same chart for men and women.
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
No of groups are 5, refer group as G1, G2, G3, G4, G5

```python
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]
index = np.arange(5)
opacity = 0.7
bar_width = 0.2

rects1 = plt.bar(index, men_means, bar_width, alpha = opacity,color = 'g', label = 'Men')

rects2 = plt.bar(index+bar_width, women_means, bar_width, alpha = opacity,color = 'r', label = 'Women')

plt.xlabel('Person')
plt.xlabel('Scores')
plt.title('Score By Person')

plt.xticks(index+0.15,('G1','G2','G3','G4','G5'))
plt.legend()
```

## Make the above graph Horizontal

```python
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]
index = np.arange(5)
opacity = 0.7
bar_width = 0.4

p1 = plt.barh(index, men_means, bar_width, color='red')
p2 = plt.barh(index, women_means, bar_width,
left =men_means, color='green')

plt.ylabel('Groups')
plt.xlabel('Scores')
plt.title('Scores by group\n' + 'and gender')
plt.yticks(index, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
```

### Create % Graph

```python
n_groups = 5

men_means = np.array(men_means)
women_means = np.array(women_means)
total = men_means + women_means #add the array elements together according to the position
proportion_men_means = np.true_divide(men_means, total) * 100 #numpy.true_divide(x1, x2).x1---> divident and x2---> divisor
proportion_women_means = np.true_divide(women_means, total) * 100

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
index = np.arange(n_groups)
ax.bar(index, proportion_men_means, bar_width, color='r')
ax.bar(index, proportion_women_means, bar_width,bottom=proportion_men_means, color='b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
plt.show()
```
