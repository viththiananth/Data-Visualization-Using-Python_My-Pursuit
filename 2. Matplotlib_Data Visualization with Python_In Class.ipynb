```python
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]
index = np.arange(5)
opacity = 0.7
bar_width = 0.4

#Plot with 'barh' Command
p1 = plt.barh(index, men_means, bar_width, color='steelblue')
p2 = plt.barh(index, women_means, bar_width, left =men_means, color='green')

#Inter-change the xlabel and ylabel
plt.ylabel('Groups')
plt.xlabel('Scores')
plt.title('Scores by group\n' + 'and gender')
plt.yticks(index, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
#plt.yticks(np.arange(0, 10, 1))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
```

### Create % Graph

```python
#Convert the List into Arrays
men_means = np.array(men_means)
women_means = np.array(women_means)
#add the array elements together according to the position
total = men_means + women_means 
#numpy.true_divide(x1, x2).x1---> divident and x2---> divisor (Get the Percentage Value)
proportion_men_means = np.true_divide(men_means, total) * 100 
proportion_women_means = np.true_divide(women_means, total) * 100

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
index = np.arange(5)

ax.bar(index, proportion_men_means, bar_width, color='r')
ax.bar(index, proportion_women_means, bar_width,bottom=proportion_men_means, color='b')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_yticks(np.arange(0, 101, 10))
ax.legend(labels=['Men', 'Women'])
plt.show()
```

## Bubble Chart

```python
#Filter out Argentina & Brazil Data
df_can = df.loc[['Argentina','Brazil']]

# transposed dataframe
df_can_t = df_can[years].transpose() 

# cast the Years (the index) to type int
df_can_t.index = map(int, df_can_t.index)

# let's label the index. This will automatically be the column name when we reset the index
df_can_t.index.name = 'Year'

# reset index to bring the Year in as a column
df_can_t.reset_index(inplace=True)
```

```python
# normalize Brazil data
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())
# normalize Argentina data
norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
# Plot Brazil Values
ax0 = df_can_t.plot(kind='scatter', x='Year', y='Brazil',figsize=(14, 8), alpha=0.5, color='green',s=norm_brazil*1000, xlim=(1975, 2014))
# Plot Argentina Values
ax1 = df_can_t.plot(kind='scatter', x='Year', y='Argentina', alpha=0.5, color="blue", s=norm_argentina *1000, ax = ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
